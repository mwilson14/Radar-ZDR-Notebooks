```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
saved_tracks_KMVX= pd.read_pickle('ExampleCase_7_11_17KMVX.pkl')
```

```python
print(saved_tracks_KMVX)
```

```python
stormA = saved_tracks_KMVX.xs(1, level=1)
stormA_alg_f = np.asarray(stormA['zdr_area'].iloc[:])
stormA_alg = stormA_alg_f

stormA = saved_tracks_KMVX.xs(1, level=1)
stormA_max_f = np.asarray(stormA['zdr_max'].iloc[:])
stormA_max = stormA_max_f

stormA = saved_tracks_KMVX.xs(1, level=1)
stormA_sep_f = np.asarray(stormA['kdp_zdr_sep'].iloc[:])
stormA_sep = stormA_sep_f

stormA_times_f = np.asarray(stormA['times'].iloc[:])
stormA_times= stormA_times_f
#Get all times as strings
timestrings = []
for i in range(len(stormA_times)):
    timestr = stormA_times[i].astype(str)
    timestr1 = timestr[11:16]
    timestrings.append(timestr1)
```

```python
print(timestrings)
```

```python
plt.style.use('bmh')
#plt.title('21 UTC NARR Sounding 31 May 1985 Wheatland, PA')
# Change default to be better for skew-T
fig = plt.figure(figsize=(24, 12))
x = np.arange(0, len(stormA_alg), 1)
#plt.plot(x, vdb_stormA, label = 'VDB 2016')
plt.plot(x, stormA_alg, label = 'Algorithm ZDR Arc Area')
#plt.plot(x[13:22], stormA_alg[13:22], color='r', label = 'ZDR Arc Area')
#plt.plot(x, stormA_alg3)
plt.xticks(x[::2], timestrings[::2], size=20, rotation=39)
plt.yticks(size=25)
plt.xlabel('Scan', size = 30)
plt.ylabel('ZDR Arc Extent (km^2)', size = 30)
plt.title('Storm 1 3.5dB ZDR Arc Area Extent', size = 35)
plt.legend(prop={'size':25})
plt.tight_layout()
#plt.savefig('ZDR_algorithm_comparison_StormA.png')
plt.show()
```

```python
plt.style.use('bmh')
#plt.title('21 UTC NARR Sounding 31 May 1985 Wheatland, PA')
# Change default to be better for skew-T
fig = plt.figure(figsize=(24, 12))
x = np.arange(0, len(stormA_alg), 1)
#plt.plot(x, vdb_stormA, label = 'VDB 2016')
plt.plot(x, stormA_max, label = 'Algorithm ZDR Arc Max Value')
#plt.plot(x, stormA_alg3)
plt.xticks(x[::2], timestrings[::2], size=20, rotation=39)
plt.yticks(size=25)
plt.xlabel('Scan', size = 30)
plt.ylabel('ZDR Arc Max Value (dB)', size = 30)
plt.title('Storm 1 3.5dB ZDR Arc Max Value', size = 35)
plt.legend(prop={'size':25})
plt.tight_layout()
#plt.savefig('ZDR_algorithm_comparison_StormA.png')
plt.show()
```
